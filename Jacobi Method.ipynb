### Nomor 1 

```python
import numpy as np #meng-import modul numpy
from tabulate import tabulate #meng-import modul tabulate

aug_matrix = np.array(eval(input('Masukkan augmented matrix dari SPL yang akan diselesaikan: '))) #untuk memasukkan matriks
initial_solution = np.array(eval(input('Masukkan tebakan awal solusi: ')),dtype = float) #untuk memasukkan tebakan awal solusi
tolerance = eval(input('Masukkan toleransi: ')) #untuk memasukkan nilai toleransi
iteration_table = [] #membuat fungsi untuk tabel iterasi

def Jacobi(matrix, initial, tol):
    n = np.shape(matrix)[0] #menentukan ukuran matriks
    x0 = initial #fungsi untuk mendefinisikan tebakan awal
    x = x0.copy() #mendefinisikan x = x0
    s = 0
    for i in range(n): 
        b = matrix[i,n] #untuk konstanta
        for j in range(n): #metode Jacobi
            if j!=i:
                b -= matrix[i,j]*x0[j]
        x[i] = b/matrix[i,i] #hasil aproksimasi
        value = x.tolist() 
    s = s+1
    value.insert(0, s)
    iteration_table.append(value)
    
    while np.linalg.norm(x - x0, np.inf) >= tol:
        x0 = x #
        x = x0.copy()
        for i in range(n):
            b = matrix[i,n]
            for j in range(n): #metode Jacobi
                if j!=i:
                    b -= matrix[i,j]*x0[j]
            x[i] = b/matrix[i,i]
            value = x.tolist()
        s = s+1
        value.insert(0, s)
        iteration_table.append(value)
    print(tabulate(iteration_table, headers = ["Iterasi", "X1", "X2", "X3", "X4"], tablefmt="fancy_grid"))
    print('Aproksimasi solusi SPL tersebut dengan metode Jacobi adalah: ')
    return x0

print(Jacobi(aug_matrix,initial_solution,tolerance))
```

**Pembahasan Nomor 1** \
\
Matriks pada nomor 1 konvergen dengan iterasi sebanyak 12 kali. \
Aproksimasi solusi dari SPL tersebut menggunakan metode Jacobi adalah [-0.75133985   0.03934973  -0.27956649   0.68930775]
